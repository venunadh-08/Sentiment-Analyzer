```python
import string
from nltk.corpus import stopwords
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))  # Move this outside the function

def text_process(msg):
    nopunc = [char for char in msg if char not in string.punctuation]
    nopunc = ''.join(nopunc)
    return [word for word in nopunc.split() if word.lower() not in stop_words]
```

```python
df['Processed Text'] = df['Content'].apply(text_process)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer,TfidfVectorizer
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.model_selection import train_test_split
```

```python
msg_train, msg_test, label_train, label_test = train_test_split(df['Content'],df['Label'],test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(analyzer=text_process)),
    ('classifier', LogisticRegression(max_iter=1000, random_state=42)),
])
```

```python
pipeline.fit(msg_train,label_train)
```

```python
pred = pipeline.predict(msg_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
print(classification_report(label_test, pred))
print(accuracy_score(label_test, pred))
```

```python
#With out pipeline
cv = CountVectorizer()
```

```python
X = df['Content']
y = df['Label']
```

```python
X = cv.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression(max_iter=1000)
```

```python
lg.fit(X_train, y_train)
```

```python
pred1 = lg.predict(X_test)
```

```python
print("With Out Pipelining")
print(classification_report(y_test, pred1))
print(accuracy_score(y_test, pred1))

print("With Pipelining")
print(classification_report(label_test, pred))
print(accuracy_score(label_test, pred))
```

```python
def predict_sentiment_p(msg):
    return pipeline.predict([msg])[0]
```

```python
def predict_sentiment(message):
    # Preprocess the input message and convert it into a feature vector
    message_vec = cv.transform([message])
    
    # Predict the sentiment using the trained model
    prediction = lg.predict(message_vec)
    
    return prediction[0]
```

```python
msg = 'Here are the details for the upcoming event.'

print(predict_sentiment_p(msg))
print(predict_sentiment(msg))
```
